### Центральная предельная теорема
В этом задании нужно убедиться, что ЦПТ действительно работает

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
rv = sts.norm(loc=1.0, scale=0.05)
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
def get_CLT_data(loc, scale, n, cases):
    plt.figure(figsize=(15, 6))
    rv = sts.norm(loc=loc, scale=scale)
    means_array = np.array([])
    for _ in range(cases):
        means_array = np.append(means_array, rv.rvs(size=n).mean())
    plt.hist(means_array, density=True, label='sample of ' + str(n) + ' histogram', bins=1000, range=(0.80, 1.20))
    plt.legend()
    plt.ylabel('samples means')
    plt.xlabel('$x$')
```

```python
get_CLT_data(1.0, 0.05, 2, 1000)
get_CLT_data(1.0, 0.05, 5, 1000)
get_CLT_data(1.0, 0.05, 10, 1000)
get_CLT_data(1.0, 0.05, 30, 1000)
get_CLT_data(1.0, 0.05, 500, 1000)
```
